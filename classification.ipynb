```python
import pandas as pd
import sys
#reload(sys)
#sys.setdefaultencoding('utf-8')
from fasttext import load_model

classifier = load_model("commits.bin") 
df = pd.read_csv('total_processed.csv',  lineterminator='\n', encoding="utf-8")
#df = df.dropna()
df = df.replace('\n','', regex=True)
#df['message'] = df['message'].astype(str)
```

```python
df.columns
```

```python
df.shape
```

```python
commits = list(df['message'].astype(str))
```

```python
labels = classifier.predict(commits)  
```

```python
res = list(zip(*labels))
```

```python
res_list = [x[0] for x in res]
```

```python
lst2 = [item[0] for item in res_list]
```

```python
df['labels_predicted'] = lst2
```

```python
y_test = list(df['results'])
y_pred = list(df['labels_predicted'])

df['category_id'] = df['results'].factorize()[0]
category_id_df = df[['results', 'category_id']].drop_duplicates().sort_values('category_id')

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

conf_mat = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(conf_mat, annot=True, fmt='d',
            xticklabels=category_id_df.results.values, yticklabels=category_id_df.results.values)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
